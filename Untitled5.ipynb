```python
import numpy as np
import wave
import math
import librosa.feature
from maliAlek.funkcije import *
from maliAlek.pomagala import *
from maliAlek.konstante import *
```

```python
raw = wave.open("./maliAlek/ulaz.wav")
# raw = wave.open("./maliAlek/untitled.wav")
signal = raw.readframes(-1)
fs = raw.getframerate()

signal = np.frombuffer(signal, dtype ="int16").astype("float64")
signal = signal[:(len(signal) // framesInSegment) * framesInSegment]

plotuj(np.linspace(0, len(signal) / sampleRate, num = len(signal)), signal)
```

```python
signal = separate(signal, framesInSegment, len(signal))
activity = [] #lista za aktivnost svakog segmenta
activeSegments = []
activeIndexes = []

for i, delic in enumerate(signal):
    time = np.linspace((i * segmentDuration // 2) / sampleRate, (i * segmentDuration // 2 + segmentDuration) / sampleRate, num = framesInSegment) * 50
    if(active(delic, activity[-60:])):
        activity.append(True)
        plotuj(time, delic) 
        activeSegments.append(delic)
        activeIndexes.append(i)
    else:
        activity.append(False)
```

```python
n_mfcc = 13
mfccs = []
mfc_Sens = 350

for i, segment in enumerate(activeSegments):
  mfccs.append(librosa.feature.mfcc(y = segment, sr = sampleRate, n_mfcc = n_mfcc, n_fft = framesInSegment).reshape(n_mfcc * 3))


def same(mfccs, x, mfc_Sens):
  vreme = np.arange(len(x)) + 1
  plotuj(vreme, x)
  res = []
  for brojac, m in enumerate(mfccs):
    diff = 0
    for i, c in enumerate(m): diff += abs(c - x[i])
    if(diff <= mfc_Sens and diff != 0): 
      # print(brojac, brojac )
      plotuj(vreme, m)
      res.append(brojac)
  
  return res


# for i, m in enumerate(mfccs):
#   r = same(mfccs, m, mfc_Sens)
#   if(r != []): print(i, r, activeIndexes[i])
r = same(mfccs, mfccs[54], mfc_Sens)
r.append(54)
r
```
